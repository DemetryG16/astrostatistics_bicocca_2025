```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

N = 100000
x1 = sp.symbols('x1')
x = np.random.uniform(0.1,10,N)
plt.hist(x, bins=60, color='blue')
plt.title('Uniform distribution')
plt.grid()
plt.show()
x1 = np.array(x)


py = x1*(np.log(10)) * 1675
# py = dx_dy * ln(10) *px // px = counts/N

y = np.log10(x)

plt.plot(x1,py)
plt.xlim(-1,1)
plt.ylim(0,8000)
plt.hist(y, bins=60, color='blue')
plt.title('Log10 distribution')
plt.show()
```

```python
print('Mean of x:', np.log10(np.mean(x)))
print('Mean of y:', np.mean(y)) 
print("The median of x is: ", np.log10(np.median(x)))
print("The median of y is: ", np.median(y))
# Practicing the derivative of a function
counts = np.histogram(x)
print(counts)
```
