```python
# L02 - Monty Hall Problem
# 1. A player chooses one of three doors
# 2. The host opens one of the other two doors, revealing a goat
# 3. The player can choose to switch to the other unopened door or stay with their original choice
import random as rand

trials=10000
player1=0 # conservative
player2=0 # Switch
player3=0 # new player
d = 3     # Doors number
for i in range(trials):
  choose = rand.randint(1,d)
  car = rand.randint(1,d)
  for door in range(1,d+1):
     if door != car and door != choose:
      open_door = door
      
  for door in range(1,d+1):
      if door != choose and door != open_door:
          switch_door= door
         
  if choose == car:
     player1 +=1
  if switch_door == car:
     player2 +=1
  player3counter= rand.randint(1,d)
  if player3counter == car:
      player3 +=1

P1 = player1/trials
P2 = player2/trials
P3 = player3/trials
print("Conservative Player =", P1)        
print("Player switch = ", P2) 
print("New Player =", P3)
```

Absolutly I choose the player who switch the choice after opening one door
